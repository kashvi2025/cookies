```python
fig, axs = plt.subplots(2, 1)
axs[0].plot(x, color='orange', linewidth = '10')
axs[0].plot(y, color='red', linestyle='--')
axs[0].set_title('top')
axs[0].set_ylabel('Values')
axs[1].scatter(range(len(z)), z, color='magenta', marker='D')
axs[1].set_title('bottom')
axs[1].set_xlabel('position')
axs[1].set_ylabel('values')
plt.show()
```
